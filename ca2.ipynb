```python
# Analysis of the correlation between numerical features of EV dataset using a full heatmap.
numeric_df = df.select_dtypes(include=np.number)
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="Purples",
    linewidths=0.5,
    square=True,
    cbar_kws={'label': 'Correlation Coefficient'}
)
plt.title("Full Correlation Heatmap of Numeric Features", fontsize=16, weight='bold')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Analyzed EV data using combo chart, bar plots for Ev registrations by Model Year with Average line
df = df[df['Model Year'].notna()]
df['Model Year'] = df['Model Year'].astype(int)
year_counts = df['Model Year'].value_counts().sort_index()
average_count = year_counts.mean()
plt.figure(figsize=(14, 7))
sns.barplot(x=year_counts.index, y=year_counts.values, color='green')
plt.axhline(y=average_count, color='red', linestyle='--', linewidth=2, label=f'Average: {average_count:.0f}')
plt.title("EV Registrations by Model Year with Average Line", fontsize=16, weight='bold')
plt.xlabel("Model Year", fontsize=12)
plt.ylabel("Number of EVs", fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Analyzed EV data using combo chart, bar plots for Average Electric Range by Model Year 
range_per_year = df.groupby('Model Year')['Electric Range'].mean().dropna().sort_index()
range_avg = range_per_year.mean()
# Plot
plt.figure(figsize=(14, 7))
sns.barplot(x=range_per_year.index, y=range_per_year.values, color='#FFA500')
# Add average line
plt.axhline(range_avg, color='darkred', linestyle='--', linewidth=2, label=f'Avg Range: {range_avg:.1f} mi')
# Titles and styling
plt.title("Average Electric Range by Model Year", fontsize=16, weight='bold')
plt.xlabel("Model Year")
plt.ylabel("Average Range (miles)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Trend Analysis of EV Registrations Over the Years Using Line Plot
df = df[df['Model Year'].notna()]
# Count number of EVs by year
yearly_counts = df['Model Year'].value_counts().sort_index().reset_index()
yearly_counts.columns = ['Model Year', 'Count']
# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_counts, x='Model Year', y='Count', marker='o', color='green')
plt.title('EV Registrations Over the Years')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Comparison of EV Registrations Across Top Makes Using Bar Charts
df = df[df['Make'].notna()]
top_makes = df['Make'].value_counts().head(10).reset_index()
top_makes.columns = ['Make', 'Count']
plt.figure(figsize=(10, 6))
sns.barplot(data=top_makes, x='Count', y='Make', hue='Make', palette='dark:red', dodge=False)
plt.title('Top 10 Most Common EV Makes')
plt.xlabel('Number of Vehicles')
plt.ylabel('Make')
plt.legend([],[], frameon=False)
plt.tight_layout()
plt.show()
```

```python
# Comparison of EV Registrations Across Top Cities Using Bar Charts"
df.columns = df.columns.str.strip()
df = df[df['City'].notna()]

# Get top 10 cities with most EV registrations
top_cities = df['City'].value_counts().head(10).reset_index()
top_cities.columns = ['City', 'Count']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_cities, x='Count', y='City', hue='City', palette='dark:red', dodge=False)
plt.title('Top 10 Cities by EV Registrations')
plt.xlabel('Number of Vehicles')
plt.ylabel('City')
plt.legend([], [], frameon=False)  
plt.tight_layout()
plt.show()
```

```python
# Analyzing Electric Range Trends Over Model Years for Top 5 EV Manufacturers Using Scatter Plot
df = df[df['Model Year'].notna() & df['Electric Range'].notna() & df['Make'].notna()]
df['Model Year'] = df['Model Year'].astype(int)
df['Electric Range'] = pd.to_numeric(df['Electric Range'], errors='coerce')

# Top 5 makes
top_makes = df['Make'].value_counts().head(5).index
filtered_df = df[df['Make'].isin(top_makes)]

# Color map for top 5 makes
colors = ['red', 'blue', 'green', 'purple', 'orange']
make_color_map = dict(zip(top_makes, colors))

# Plotting
plt.figure(figsize=(12, 6))
for make in top_makes:
    subset = filtered_df[filtered_df['Make'] == make]
    plt.scatter(
        subset['Model Year'],
        subset['Electric Range'],
        label=make,
        color=make_color_map[make],
        edgecolors='black',
        alpha=0.6,
        s=60
    )

# Styling
plt.title('Electric Range vs Model Year (Top 5 EV Makes)', fontsize=16, fontweight='bold')
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Electric Range (miles)', fontsize=12)
plt.legend(title='Make', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Distribution of Clean Alternative Fuel Vehicle (CAFV) Eligibility Using Pie Chart
df = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].notna()]
cafv_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
plt.figure(figsize=(8, 6))
cafv_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140,
                 colors=['#ffb703', '#219ebc', '#fb8500'])
plt.ylabel('')
plt.title('CAFV Eligibility Distribution')
plt.tight_layout()
plt.show()
```

```python
# Electric Range Comparison by Vehicle Type Using Box Plot
df = df[df['Electric Vehicle Type'].notna()]
df = df[df['Electric Range'].notna()]
# Plot with hue and no legend to avoid FutureWarning
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Electric Vehicle Type', y='Electric Range',
            hue='Electric Vehicle Type', palette='Set2', dodge=False)
plt.title('Electric Range by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Electric Range (miles)')
plt.legend([], [], frameon=False) 
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
